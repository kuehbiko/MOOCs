### Question 2:  Remove STOP words* 

    *In NLP, stop words are commonly used words like "a", "is", and "the". 
    They are typically filtered out during processing.

<b>a. Implement a function that takes a string text and an integer k. </b>

<b>b. Checks the occurrence of each word in the sentence. </b>

<b>c. Returns the list of words that occur in the text at least k times. </b>

<b>d. The words must be returned in the order of their first occurrence in the text. </b>

##### Examples:

input:

    text = "a mouse is smaller than a dog but a dog is stronger"
    k=2

output: 

    ['a', 'is', 'dog']

##### Explanation: 
- The list of stop words that occur at least k = 2 times is ["a", "is", "dog"]. 
- "a" occurs 3 times, "is" and "dog" both occur 2 times. No other word occurs at least 2 times. 
- The answer is in order of the first appearance in text.




```python
# Question 2: insert your code here
def remove_stop(s = input("Enter String: "), k = input("Enter Integer: ")):
    
    # define variables
    s_lst = s.lower().split() # convert to list
    k = int(k) # assumption that input k is always a digit
    
    # count occurrence of each word
    s_dict = {key: s_lst.count(key) for key in s_lst}
    
    # get list of words that occur at least k times
    stop_words = [key for key, val in s_dict.items() if val >= k]
    
    return stop_words

remove_stop()
```

### Question 3: Suffix stripping Stemmer* 

    *Stemming is the process of extracting the base word from a word. For instance, the base for "worked" is "work".
    
<b>a) Implement a function that takes a sentence without punctuation. </b>

<b>b) If the word ends in 'ed', 'ly' or 'ing', remove the suffix. </b>

<b>c) If the resulting word is longer than 7 letters, keep the first 7 letters. </b>

<b>d) Return the stemmed counterpart of individual words. </b>

##### Example:

input:

    text = "an extremely ferocious dog is barking"

output: 

    an extreme ferocio dog is bark

##### Explanation: 
- 'an' does not end in one of the suffixes and is less than 8 letters
- 'extremely' is 'extreme' after removing the suffix and 'extreme' is less than 8 letters
- 'ferocious' is 9 letters long, so reduce it to 7 letters
- ' ferocio','dog' and 'is' are unchanged
- 'barking' is 'bark' after removing the suffix and is less than 7 letters

```python
# Question 3: insert your code here
def stem_stripper(s = input("Enter a sentence")):

    # define variables (assume input sentence already has no punctuation)
    s_list = s.split()
    suffixes = ['ed', 'ly', 'ing']
    
    # 1. remove suffixes
    def remove_suffix(w, suffixes):
        for suffix in suffixes:
            if w.endswith(suffix): return w[:-len(suffix)]
        return w
    s_list = [remove_suffix(w, suffixes) for w in s_list]
    
    #s_list = [lambda w,suf : w[:-len(suf)] if w.endswith(suf) else w for w in s_list and suf in suffixes]
    
    # 2. remove 8th+ letters
    s_list = [w[:8] if len(w)>7 else w for w in s_list]
    
    return ' '.join(s_list)
    
stem_stripper() 
```

```python
#thanks chatgpt
def scientific_calculator(operation, num1, num2=None):
    operations = {
        'add': lambda x, y: x + y,
        'subtract': lambda x, y: x - y,
        'multiply': lambda x, y: x * y,
        'divide': lambda x, y: "Cannot divide by zero" if y == 0 else x / y,
        'exponential': lambda x, y: x ** y,
        'sqrt': lambda x, y=None: "Cannot compute square root of a negative number" if x < 0 else math.sqrt(x),
        'log': lambda x, y: "Arguments must be greater than zero for logarithm" if x <= 0 or y <= 0 else math.log(x, y),
        'sin': lambda x, y=None: math.sin(x),
        'cos': lambda x, y=None: math.cos(x),
        'tan': lambda x, y=None: math.tan(x)
    }

    if operation in operations:
        return operations[operation](num1, num2)
    else:
        return "Invalid Operation"
```
