# 1. Hierarchical Clustering
    
## What is Hierarchical Clustering? <a id='p2.1' />

Hierarchical clustering involves creating clusters that have a predetermined ordering from top to bottom. For example, all files and folders on the hard disk are organized in a hierarchy

Imagine you have some number of clusters k you are interested in finding. All you know is that you can probably break up your dataset into that many distinct groups at the top level, but you might also be interested in the groups inside your groups, or the groups inside of those groups. To get that kind of structure, we use hierarchical clustering.

We begin with n different points and k different clusters we want to discover; for our purposes, n = 4, and k = 2.

<img src = "images/hierarchial1.png" height = "400" width = "400"><br><br>

Start by treating each point as if it were its own cluster.

<img src = "images/hierarchial2.png" height = "400" width = "400"><br><br>

We then start merging each of our single-point clusters into larger clusters, using the clusters that are closest together. We find the smallest distance from the pairwise distance matrix‚Ää‚Äî‚Ääwhich is just a table of the distance of every cluster from every other cluster.

This is initialized as the Euclidean distance between each point, but after that, we switch to one of a variety of different ways of measuring cluster distance.

We now start merging the clusters that are closest together. Say we know that x1 and x3 are closest together. We then merge those two into a new cluster, ca.

<img src = "images/hierarchial3.png" height = "400" width = "400"><br><br>

We now recalculate ca‚Äôs distance from every other cluster, using whichever method we prefer from the aforementioned grab-bag. Then we repeat, merging our clusters over and over until we get k top-level clusters‚Ää‚Äî‚Ääin our example, two clusters. Say we figure out that x2 is closer to ca than to x4.

<img src = "images/hierarchial4.png" height = "400" width = "400"><br><br>

We now have two top-level clusters, cb and x4 (remember that each point starts as its own cluster). We can now search the tree structure we‚Äôve created to search for sub-clusters, which in our original 2-D view would look like this:

<img src = "images/hierarchial5.png" height = "400" width = "400"><br><br>

The main output of Hierarchical Clustering is a dendrogram, which shows the hierarchical relationship between the clusters:

<img src = "images/hierarchial_dendogram.png" height = "400" width = "400"><br><br>

The key to interpreting a dendrogram is to focus on the height at which any two objects are joined together. In the example above, we can see that E and F are most similar, as the height of the link that joins them together is the smallest. The next two most similar objects are A and B.

Observations are allocated to clusters by drawing a horizontal line through the dendrogram. Observations that are joined together below the line are in clusters. In the example below, we have two clusters, one that combines A and B, and a second combining C, D, E, and F.

<img src = "images/dendogram2.png" height = "400" width = "400"><br><br>

### Types of Hierarchical Clustering <a id='p2.1.1' />

<img src = "images/hctypes.png">

<font size = "3"><b>Divisive method:</b></font><br>
In divisive or top-down clustering method we assign all of the observations to a single cluster and then partition the cluster to two least similar clusters. Finally, we proceed recursively on each cluster until there is one cluster for each observation. There is evidence that divisive algorithms produce more accurate hierarchies than agglomerative  algorithms in some circumstances but is conceptually more complex.		
 		
<font size = "3"><b>Agglomerative method:</b></font><br>
In agglomerative or bottom-up clustering method we assign each observation to its own cluster. Then, compute the similarity (e.g., distance) between each of the clusters and join the two most similar clusters. Finally, repeat steps 2 and 3 until there is only a single cluster left.<br>
***

### Types of Linkages <a id='p2.1.2' />

Before any clustering is performed, it is required to determine the proximity matrix containing the distance between each point using a distance function. Then, the matrix is updated to display the distance between each cluster. The following methods differ in how the distance between each cluster is measured.
Different types of linkages are:
* Single linkage
* Complete linkage
* Average linkage
* Ward's linkage
* Centroid linkage<br>
***

<font size = "4"><b>Single Linkage</b></font>

In single linkage hierarchical clustering, the distance between two clusters is defined as the shortest distance between two points in each cluster.<br>
<mark>For example, the distance between clusters ‚Äúr‚Äù and ‚Äús‚Äù to the left is equal to the length of the arrow between their two closest points.</mark>

<img src="images/singlelinkage.png">


<font size = "4"><b>Complete Linkage</b></font>

In complete linkage hierarchical clustering, the distance between two clusters is defined as the longest distance between two points in each cluster.<br>
<mark>For example, the distance between clusters ‚Äúr‚Äù and ‚Äús‚Äù to the left is equal to the length of the arrow between their two furthest points.</mark>

<img src="images/completelinkage.png">

		
<font size = "4"><b>Average Linkage</b></font>

In average linkage hierarchical clustering, the distance between two clusters is defined as the average distance between each point in one cluster to every point in the other cluster.<br>
<mark>For example, the distance between clusters ‚Äúr‚Äù and ‚Äús‚Äù to the left is equal to the average length each arrow between connecting the points of one cluster to the other.</mark>

<img src="images/averagelinkage.png">

<font size = "4"><b>Ward's Linkage</b></font>

Assumes that a cluster is represented by its centroid,
and measures the proximity between two clusters in
terms of the increase in sum of the squared error (SSE)
that results from merging the two clusters:

$ùëë(ùê¥, ùêµ) = ùëÜùëÜùê∏ (ùê¥ U ùêµ ) ‚àí ùëÜùëÜùê∏_ùê¥ ‚àí ùëÜùëÜùê∏_ùêµ$

where ùê¥ and ùêµ are clusters.<br>
Note that for hierarchical clustering, the SSE starts at 0<br>


<font size = "4"><b>Centroid Linkage</b></font>

This involves finding the mean vector location for each of the clusters and taking the distance between the two centroids.

### Importing all the essential libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from sklearn import metrics # for evaluations
from sklearn.datasets import make_blobs, make_circles # for generating experimental data
import cmath as math
import matplotlib.cm as cm
import sys
```

<font style="font-family:Trebuchet MS;">

### Generating an ideal dataset that can help us understand the basic implementation <a id='p2.4.2' />
*Note that in practical cases the data will not be this ideal*

<mark style="background-color:#ADD8E6;">random_state: This allows the reproducibility of data</mark>

```python
# This particular setting has one distinct cluster and 4 clusters placed close together.
X, y = make_blobs(n_samples=500,
                  n_features=2,
                  centers=4,
                  cluster_std=1,
                  center_box=(-10.0, 10.0),
                  shuffle=True,
                  random_state=1)
# Plot the data distribution.
plt.scatter(X[:,0], X[:,1], c=y)
```

<font style="font-family:Trebuchet MS;">

### Creating the Dendrogram as it helps visualize the relationship between similar sets of data <a id='p2.4.3' />

```python
"""
Creating the dendrogram

Parameters
----------

method: {'single', 'complete', 'average'}
        This represents the kind of linkage algorithm to be used

"""

mergings = linkage(X, method='ward')
dendrogram(mergings)
plt.show()
```

<font style="font-family:Trebuchet MS;">

### Selecting the number of clusters and analyzing cluster quality with silhouette analysis <a id='p2.4.4' />
 


```python
# List of number of clusters
range_n_clusters = [2, 3, 4]

# We have to create a dictionary mapping the number of cluster to the appropriate thershold (t) in the dendrogram
# Threshold is required to assign the clusters as it tells the algorithm where to cut the dendrogram
dendogram_to_clusters = {2:140,3:50,4:20}

# For each number of clusters, perform Silhouette analysis and visualize the results.
for n_clusters in range_n_clusters:
    
    # Perform hierarchical clustering
    mergings = linkage(X, method='ward')
    y_pred = fcluster(mergings, dendogram_to_clusters[n_clusters], criterion='distance')
    
    # Compute the cluster homogeneity and completeness.
    homogeneity = metrics.homogeneity_score(y, y_pred)
    completeness = metrics.completeness_score(y, y_pred)
    
    # Compute the Silhouette Coefficient for each sample.
    s = metrics.silhouette_samples(X, y_pred)
    
    # Compute the mean Silhouette Coefficient of all data points.
    s_mean = metrics.silhouette_score(X, y_pred)
    
    # For plot configuration -----------------------------------------------------------------------------------
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)
    
    # Configure plot.
    plt.suptitle('Silhouette analysis for hierarchical clustering with n_clusters: {}'.format(n_clusters),
                 fontsize=14, fontweight='bold')
    
    # Configure 1st subplot.
    ax1.set_title('Silhouette Coefficient for each sample')
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")
    ax1.set_xlim([-1, 1])
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])
    
    # Configure 2st subplot.
    ax2.set_title('Homogeneity: {}, Completeness: {}, Mean Silhouette score: {}'.format(homogeneity,
                                                                                        completeness,
                                                                                        s_mean))
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")
    
    # For 1st subplot ------------------------------------------------------------------------------------------
    
    # Plot Silhouette Coefficient for each sample
    y_lower = 10
    for i in range(n_clusters):
        i=i+1
        ith_s = s[y_pred == i]
        ith_s.sort()
        size_cluster_i = ith_s.shape[0]
        y_upper = y_lower + size_cluster_i
        cmap = cm.get_cmap("Spectral")
        color = cmap(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_s,
                          facecolor=color, edgecolor=color, alpha=0.7)
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
        y_lower = y_upper + 10
        
    # Plot the mean Silhouette Coefficient using red vertical dash line.
    ax1.axvline(x=s_mean, color="red", linestyle="--")
    
    # For 2st subplot -------------------------------------------------------------------------------------------
    
    # Plot the predictions
    cmap = cm.get_cmap("Spectral")
    colors = cmap(y_pred.astype(float) / n_clusters)
    ax2.scatter(X[:,0], X[:,1], c=colors)
```

<font style="font-family:Trebuchet MS;">
The silhouette plot shows that the n_clusters value of 3 is a bad pick for the given data due to the presence of clusters with below average silhouette scores and also due to wide fluctuations in the size of the silhouette plots. Silhouette analysis is more ambivalent in deciding between 2 and 4.<BR>

Also from the thickness of the silhouette plot the cluster size can be visualized. The silhouette plot for cluster 2 when n_clusters is equal to 2, is bigger in size than cluster 1. However when the n_clusters is equal to 4, all the plots are more or less of similar thickness and hence are of similar sizes as can be also verified from the labeled scatter plot on the right. <b>So, we should go ahead with n_clusters equal to 4.</b>

<font style="font-family:Trebuchet MS;">
    
### Assigning a cluster and using a cross tab to analyze the number of data points assigned to clusters <a id='p2.4.5' />

<mark style="background-color:#ADD8E6;">Check the dendrogram and set the parameter $t$ to a value (say t = x) such that when we draw a horizontal line with $y = x$ in the dendrogram the number of intersection points will be equal to the number of clusters we want</mark>

```python
# Since we have decided to go with 4 clusters, the threshold (t) that the dendrogram has to be cut at is 20 to achieve this
# This can be seen from the dendrogram plotted above
Clusters = fcluster(mergings, t = 20, criterion='distance',)
df = pd.DataFrame({'Clusters': Clusters, 'Y': y})
ct = pd.crosstab(df['Clusters'], df['Y'])
ct
```

<font style="font-family:Trebuchet MS;">
    
### Calculating the external evaluation metrics <a id='p2.4.6' />

```python
def Evaluate(df):
    ent = {}
    tp = ct.apply(max, axis = 1).sum()
    n = ct.sum().sum()
    p = (tp/n) 
    for i in ct.index.values:
        l1 = list(ct.iloc[i-1,:])
        l2 = [(j*np.log(j/sum(l1))/sum(l1)) for j in l1 if j!=0]
        ent[i] = -sum(l2)

    entropy = sum([(ent[i]*sum(ct.iloc[i-1,:]))/n for i in ent])
    return (p,entropy)
```

## Dimensionality Reduction


<br>
<div style="text-align: center"><font size="3"><font style="font-family:Trebuchet MS;"><mark>"Brevity is the soul of wit"</mark></font></font></div> 

This powerful quote by William Shakespeare can be applied to data science & analytics as well through **dimensionality reduction**

In machine learning classification problems, there are often too many features on the basis of which the final model is built. The higher the number of features, the harder it gets to visualize the training set and then work on it. Sometimes, most of these features are correlated, and hence redundant. This is where dimensionality reduction algorithms come into play.

Dimensionality reduction is the process of reducing the number of random variables under consideration, by obtaining a set of principal variables.

The different types of dimensionality reduction techniques are:
* PCA: Principal Component Analysis
* Random Projections
* Feature Agglomeration

*Here we will be talking only about Feature Agglomeration as it employs Hierarchical Clustering*

<font style="font-family:Trebuchet MS;">
    
### Introduction to Feature Agglomeration <a id='p2.5.1' />

* Feature Agglomeration uses Hierarchical Clustering (Agglomerative Clustering) to cluster similar features together
* This approach implements clustering in the feature direction, in other words clustering the transposed data
* This clusters the similar features together
* The similar features are then combined together through a pooling function which can be passed as the parameter of the feature agglomeration function in Python

Function used: **sklearn.cluster.FeatureAgglomeration**

<font style="font-family:Trebuchet MS;">
    
### Implementation of Feature Agglomeration <a id='p2.5.2' />

```python
# Importing the essential packages

import numpy as np
import pandas as pd
from sklearn.datasets import load_iris #importing the iris dataset from the pre-stored datasets under scikit learn
from sklearn.cluster import FeatureAgglomeration #package to be used for Feature Agglomeration
import matplotlib.pyplot as plt
```

```python
# Loading the iris dataset

iris = load_iris()
X = iris["data"]
Y = iris["target"]
```

```python
"""
Implementing the Feature Agglomeration function to reduce the number of features in the existing dataset from 4 to 2

Parameters:

n_clusters : int, default 2
            The number of clusters to find or the number of features in the final dataset

linkage : {‚Äúward‚Äù, ‚Äúcomplete‚Äù, ‚Äúaverage‚Äù, ‚Äúsingle‚Äù}, optional (default=‚Äùward‚Äù)
            Which linkage criterion to use. The linkage criterion determines which distance to use between sets of features.
            The algorithm will merge the pairs of cluster that minimize this criterion.

pooling_func : callable, default np.mean
            This combines the values of agglomerated features into a single value

"""
agglo=FeatureAgglomeration(n_clusters=2).fit_transform(X)
```

```python
# Plotting the datapoints on the new feature space made up of the 2 new features

plt.scatter(agglo[:,0],agglo[:,1],c=Y)
plt.show()
```

# 2. DBSCAN: Density-Based Spatial Clustering of Applications with Noise
    
## What is DBSCAN? <a id='p3.1' />

DBSCAN is a clustering method that is used in machine learning to separate clusters of high density from clusters of low density

This approach requires two parameters: a radius ùúñ and a neighborhood density ùõ≥. For every point, we calculate Neps(x)‚Ää‚Äî‚Ääthe number of points that are at most ùúñ away from x. If Neps(x) ‚â• ùõ≥, not counting x, then x is considered a core point. If a point isn‚Äôt a core point, but it is a member of a core point‚Äôs neighborhood, then it is considered a border point. Everything else is just considered noise.

One of the most common and, indeed, performative implementations of density-based clustering is Density-based Spatial Clustering of Applications with Noise, better known as DBSCAN. DBSCAN works by running a connected components algorithm across the different core points. If two core points share border points, or a core point is a border point in another core point‚Äôs neighborhood, then they‚Äôre part of the same connected component, which forms a cluster.

Say we have a relatively small ùúñ and a decently large ùõ≥. We might wind up with clusters like this:

<img src = "images/densitybased.png" height = "400" width = "400">


See? Those two lonely points are very far away from the two clusters, and there are too few for them to really mean anything‚Ää‚Äî‚Ääso they‚Äôre just noise.

Notice that we‚Äôre also able to discover non-convex clusters this way (see the orange arc). Pretty neat!

We don‚Äôt need to specify a number of clusters we‚Äôre interested in for density-based clustering to work‚Ää‚Äî‚Ääit will automatically discover some number of clusters based on your ùúñ and ùõ≥. This is especially useful when you expect all of your clusters to have a similar density.

### Parameters <a id='p3.1.1' />
- $Eps$: Maximum radius of the neighborhood.
- $MinPts$: Minimum number of points in the Eps-neighborhood of a point.

### Terms <a id='p3.1.2' />
- The Eps-neighborhood of a point $q$Ôºç$N_{Eps}$: A point $p \in N_{Eps}(q)$ if $D(p,q) \leq Eps$. (Point inside the circle).
- Outlier: Not in a cluster.
- Core point: $\left\vert N_{Eps}(q) \right\vert \geq MinPts$ (dense neighborhood).
- Border point: In cluster but neighborhood is not dense.

<div style="text-align:center"><img width="300px" src="images/core-border-point.png"/></div>


- Directly density-reachable: A point $p$ is **directly density-reachable** from a point $q$ w.r.t $Eps$ and $MinPts$ if:
    - $p \in N_{Eps}(q)$, and $q$ is a **core point**.
    - $p$ **doesn't** need to be a core point.

<div style="text-align:center"><img width="250px" src="images/directly-density-reachable.png"/></div>

- Density-reachable: A point $p$ is **density-reachable** from a point $q$ w.r.t. $Eps$ and $MinPts$ if there is a chain of points $p_1, \dots, p_n,\ p_1 = q,\ p_n = p$ such that $p_{i+1}$ is directly density-reachable from $p_i$.

<div style="text-align:center"><img width="150px" src="images/density-reachable.png"/></div>

### The Algorithm <a id='p3.1.3' />
1. Randomly choose a point $p$.
2. Retrieve all points density-reachable from $p$ w.r.t. $Eps$ and $MinPts$.
3. If $p$ is a core point, a cluster is formed.
4. If $p$ is a border point, no points are density-reachable from $p$, then visit the next point.
5. Repeat the process until all the data points have been processed.

</font>

<font style="font-family:Trebuchet MS;">

## Why do we use DBSCAN? <a id='p3.2' />

* Hierarchical clustering is advantageous for understanding any hidden structures in your data, but it has a major pitfall. In assumes that every data point is relevant‚Ääwhich is almost never the case in the real world.
* Similarly, K-Means also has no notion of outliers and it assigns all the points to a cluster even if they do not belong to any.
* Density-based clustering methods provide a safety valve. Instead of assuming that every point is part of some cluster, we only look at points that are tightly packed and assume everything else is noise.

## When do we use DBSCAN?

* When there are a lot of outliers in the data
* When we do not know the number of cluster beforehand

Let's begin to perform *DBSCAN*

<font style="font-family:Trebuchet MS;"> <a id='p3.4' />

## How to implement DBSCAN?

<font style="font-family:Trebuchet MS;">
    
### Importing essential packages <a id='p3.4.1' />

```python
# Start from importing necessary packages.
import warnings
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from IPython.display import display
from sklearn import metrics # for evaluations
from sklearn.datasets import make_blobs, make_circles # for generating experimental data
from sklearn.preprocessing import StandardScaler # for feature scaling
from sklearn.cluster import DBSCAN

# make matplotlib plot inline (Only in Ipython).
warnings.filterwarnings('ignore')
%matplotlib inline
```

<font style="font-family:Trebuchet MS;">
    
### Generating an ideal dataset <a id='p3.4.2' />

```python
# Generate data.
# This particular setting has one distinct cluster and 4 clusters placed close together.
X, y = make_blobs(n_samples=500,
                  n_features=2,
                  centers=4,
                  cluster_std=1,
                  center_box=(-10.0, 10.0),
                  shuffle=True,
                  random_state=1)

# Plot the data distribution.
plt.scatter(X[:,0], X[:,1], c=y)
```

<font style="font-family:Trebuchet MS;">
    
### Assigning the clusters <a id='p3.4.3' />

```python
# Standardize features to zero mean and unit variance.
X = StandardScaler().fit_transform(X)

# Perform DBSCAN on the data
y_pred = DBSCAN(eps=0.3, min_samples=30).fit_predict(X)

# Plot the predictions
plt.scatter(X[:,0], X[:,1], c=y_pred)

# Print the evaluations
print('Number of clusters: {}'.format(len(set(y_pred[np.where(y_pred != -1)]))))
print('Homogeneity: {}'.format(metrics.homogeneity_score(y, y_pred)))
print('Completeness: {}'.format(metrics.completeness_score(y, y_pred)))
```

<i>The **black** data points denote the **outliers** in the above result.

Note that we don't need to specify the number of clusters with *DBSCAN* algorithm. Besides, *DBSCAN* is good at finding out the outliers without requiring some hacks like we did above in *K-means* section.

<i>Now, let's try *DBSCAN* on non-spherical data.

```python
# Generate non-spherical data.
X, y = make_circles(n_samples=1000, factor=0.3, noise=0.1)

# Standardize features to zero mean and unit variance.
X = StandardScaler().fit_transform(X)

# Perform DBSCAN on the data
y_pred = DBSCAN(eps=0.3, min_samples=10).fit_predict(X)

# Plot the data distribution. (Here's another way to plot scatter graph)
plt.plot(X[y_pred == 0, 0], X[y_pred == 0, 1], 'ro')
plt.plot(X[y_pred == 1, 0], X[y_pred == 1, 1], 'go')

# Print the evaluations
print('Number of clusters: {}'.format(len(set(y_pred[np.where(y_pred != -1)]))))
print('Homogeneity: {}'.format(metrics.homogeneity_score(y, y_pred)))
print('Completeness: {}'.format(metrics.completeness_score(y, y_pred)))
print('Mean Silhouette score: {}'.format(metrics.silhouette_score(X, y_pred)))
```

Comparing to *K-means*, we can directly apply *DBSCAN* on this form of data distribution due to the density-based clustering criterion.

Note: It's worth mention that the *Silhouette score* is generally higher for **convex** clusters than other concepts of clusters, such as density based clusters.

## Drawbacks of DBSCAN

* It doesn‚Äôt perform as well as others when the clusters are of varying density. This is because the setting of the distance threshold ùê∏ùëùùë† and minPoints for identifying the neighborhood points will vary from cluster to cluster when the density varies<br>
* This drawback also occurs with very high-dimensional data since again the distance threshold ùê∏ùëùùë† becomes challenging to estimate
* Sensitive to parameters ùê∏ùëùùë† and minPoints
* Sampling affects density measures thus impacting the clustering
